# Figures for final presentation

```python
import os

import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
from IPython.display import display, Markdown
from scipy.interpolate import interp1d
import numpy as np

from ptinsight.common.latency import calculate_latencies
from ptinsight.common.plotting import plot_init, plot_format, plot_subplot, plot_save, add_relative_time

plot_init()
```

```python
def plot_time_domains():
    fig, axs = plot_subplot(figsize=[8, 8])
    ax=axs[0]

    ax.set_xticks(range(0, 6))
    ax.set_xticklabels(["12:00", "12:01", "12:02", "12:03", "12:04", "12:05"])
    ax.set_ylabel("Ingestion Time")
    ax.set_yticks(range(0, 6))
    ax.set_yticklabels(["12:00", "12:01", "12:02", "12:03", "12:04", "12:05"])
    ax.set_xlabel("Event Time")
    ax.set_aspect("equal")
    ax.set_ylim([0, 5])
    ax.set_xlim([0, 5])
    
    ax.axvspan(1, 4, alpha=0.5, color='#0067B3')
#     ax.axhspan(1, 4, alpha=0.5, color='#0067B3')
    
    plot_format(fig)
    plot_save("time_domains_event_window")

plot_time_domains()
```

```python
def plot_empty_timeline():
    fig, axs = plot_subplot(figsize=[12, 3])
    ax=axs[0]

    ax.set_xticks(range(0, 16))
    ax.set_xticklabels([f"12:{x:02d}" for x in range(0, 16)])
    ax.set_yticks([])
    ax.set_xlim([0, 15])
    ax.set_ylim([0, 5])
    ax.set_xlabel("Event Time")
    
    plot_format(fig)
    plot_save("empty_timeline")

plot_empty_timeline()
```

```python
def plot_emergency_stop():
    fig, ax = plt.subplots(1, 1, sharex="col", figsize=[10, 4], constrained_layout=True)
    
    t1 = np.linspace(0, 10, 11)
    t2 = np.linspace(0, 10, 300)
    vs = [1.33 * x for x in [15, 15, 15, 14, 9, 3, 0.9, 0.25, 0, 0, 0]]
    ax.plot(t2, interp1d(t1, vs, kind='quadratic')(t2), linewidth = 2)
    ax.scatter(t1, vs, marker="o")

    ax.set_ylabel("Velocity [m/s]")
    ax.set_xlabel("Time [s]")
    ax.set_xlim([0, 10])
    
    plot_format(fig)
    plot_save("emergency_stop")
    
plot_emergency_stop()
```
